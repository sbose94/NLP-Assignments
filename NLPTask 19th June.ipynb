```python
from urllib import request
url = "https://www.d.umn.edu/~rmaclin/gettysburg-address.html"
response = request.urlopen(url)
rawdata = response.read().decode('utf8')
```

```python
len(rawdata)
```

```python
import nltk, re
from nltk import word_tokenize
```

```python

tokens = word_tokenize(rawdata)
len(tokens)
```

```python
tokens[10:]
```

```python
from nltk.corpus import stopwords
```

```python
nltk.FreqDist(tokens)
```

```python
stop_words = stopwords.words('english')
add_to_stop_words=[',','.','-',';',':','--',"'",'(',')','<','>']
stop_words.extend(add_to_stop_words)
stop_words=set(stop_words)
```

```python
stop_words
```

```python
for i in stop_words:
    if i in tokens:
        while i in tokens:
            tokens.remove(i)
```

```python

nltk.FreqDist(tokens)
```

```python
print("Count of words before removing sw:",len(rawdata))
print("Count of words before removing sw:",len(tokens))
```
